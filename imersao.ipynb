## Gráficos

```python
ratings_above_4_5 = dados[dados['rating'] > 4.5]
rating_counts = ratings_above_4_5['rating'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Contagem de Avaliações Acima de 4.5')
plt.xlabel('Avaliação')
plt.ylabel('Contagem')
plt.show()
```

```python
lowest_300_rupia = dados[dados['actual_price'] <= 300]
biggest_300_lowest_1500_rupia = dados[(dados['actual_price'] > 300) & (dados['actual_price'] <= 1500)]
biggest_1500_rupia = dados[dados['actual_price'] > 1500]

count_lowest_300_rupia = len(lowest_300_rupia)
count_biggest_300_lowest_1500_rupia = len(biggest_300_lowest_1500_rupia)
count_biggest_1500_rupia = len(biggest_1500_rupia)

counts = [count_lowest_300_rupia, count_biggest_300_lowest_1500_rupia, count_biggest_1500_rupia]

# Crie uma lista de cores com base nas faixas de valor
colors = ['green' if count_lowest_300_rupia > 0 else 'gray',
          'yellow' if count_biggest_300_lowest_1500_rupia > 0 else 'gray',
          'red' if count_biggest_1500_rupia > 0 else 'gray']

# Crie uma lista de rótulos para as fatias com base nas faixas de valor
labels = ['Até 300 ₹' if count_lowest_300_rupia > 0 else '',
          'Entre 300 ₹ e 1500 ₹' if count_biggest_300_lowest_1500_rupia > 0 else '',
          'Acima de 1500 ₹' if count_biggest_1500_rupia > 0 else '']

explode = (0.1, 0, 0)

# Criando o gráfico em pizza
plt.figure(figsize=(8,8))
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuição de Produtos por Faixa de Valor')
plt.axis('equal')
plt.show()
```

```python
# Organizando Category
# Dividir a coluna "Category" em várias partes com base em "|"
category_parts = dados['category'].str.split('|')

# Extrair o "category_geral" da primeira parte
dados['category_geral'] = category_parts.str[0]

# Extrair o "subCategory" da segunda parte, se existir
dados['sub_category'] = category_parts.apply(lambda x: x[1] if len(x) > 1 else None)

dados.head()
```

```python
# Supondo que você já tenha os dados carregados em um DataFrame chamado "dados"

# Gráfico de barras para a distribuição de categorias
plt.figure(figsize=(12, 6))
sns.countplot(data=dados, y='category_geral', order=dados['category_geral'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Category Geral')
plt.ylabel('Contagem')
plt.title('Distribuição de Categorias Gerais')
plt.show()

# Gráfico de barras para a distribuição de subcategorias
plt.figure(figsize=(12, 10))
sns.countplot(data=dados, y='sub_category', order=dados['sub_category'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('sub_category')
plt.ylabel('Contagem')
plt.title('Distribuição de Subcategorias')
plt.show()
```

```python
# Agrupar por subcategoria e calcular a média de preços
agregado = dados.groupby(['sub_category']).agg({'actual_price': 'mean'}).reset_index()

# Ordenar os dados pela média de preços (do maior para o menor)
agregado = agregado.sort_values(by='actual_price', ascending=False)

agregado = agregado.iloc[::-1]

# Definir os limites superior e inferior para mapear as cores
limite_inferior = 300  # Cor azul
limite_superior = 5000  # Cor vermelha

# Função para mapear um valor para uma cor com base nos limites
def mapear_cor(valor):
    norm = (valor - limite_inferior) / (limite_superior - limite_inferior)
    return plt.cm.RdYlBu_r(norm)  # Usando uma paleta de cores RdYlBu invertida

# Mapear as cores com base nas médias de preços
cores_mapeadas = [mapear_cor(preco) for preco in agregado['actual_price']]

# Criar um gráfico de barras vertical com as cores mapeadas
plt.figure(figsize=(8, 12))  # Ajuste as dimensões para a orientação vertical
barras = plt.barh(agregado['sub_category'], agregado['actual_price'], color=cores_mapeadas)  # Use barh para orientação vertical

# Criar uma barra de cores para indicar a escala de preços
sm = plt.cm.ScalarMappable(cmap=plt.cm.RdYlBu_r, norm=plt.Normalize(vmin=limite_inferior, vmax=limite_superior))
sm._A = []  # Configurar para evitar erros de escala de cores
cbar = plt.colorbar(sm)
cbar.set_label('Preço Médio em ₹')

plt.xlabel('Preço Médio em ₹')  # Ajuste os rótulos dos eixos
plt.ylabel('Subcategoria')
plt.title('Preço Médio de Produtos por Subcategoria em Rúpias (Cores de Azul a Vermelho)')
plt.show()
```
